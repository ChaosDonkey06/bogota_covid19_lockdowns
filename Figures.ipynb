```python
import matplotlib.pyplot as plt
import pandas as pd
import os

cases_df = pd.read_csv(os.path.join('data', 'cases_prepared.csv'), parse_dates=['date'])
cases_df = cases_df.groupby(['date']).sum().reset_index()

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(cases_df['date'].iloc[:60], cases_df['confirm'].iloc[:60], label='Cases')
```

```python
cases_df = cases_df.sort_values('date').set_index("date").resample("W-Sun").sum().reset_index()

fig, ax = plt.subplots(figsize=(12.5, 6))
ax.plot(cases_df['date'].iloc[:12], cases_df['confirm'].iloc[:12]/7, label='Cases')
```
